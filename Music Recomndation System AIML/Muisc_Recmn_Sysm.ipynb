```python
import pandas as pd
music=pd.read_csv('music.csv')
music
```

```python
music.drop("Unnamed: 0",axis=1,inplace=True)
#artist_name
#track_name
#release_date
#genre
music.drop("lyrics",axis=1,inplace=True)
#len
```

```python
music.dtypes
```

```python
x=music.drop('track_name',axis=1)
y=music["track_name"]
```

```python
y
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
categorical_features=["artist_name","topic","release_date","genre","len"]
one_hot=OneHotEncoder()
transformer=ColumnTransformer([("one_hot",one_hot,categorical_features)],remainder="passthrough")
transformed_x=transformer.fit_transform(x)
transformed_x

```

```python
pd.DataFrame(transformed_x)
```

```python
transformed_y = pd.get_dummies(music[["track_name"]])
transformed_y #only possible for string values
```

```python
import numpy as np
np.random.seed(42)
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
model=RandomForestRegressor()
x_train,x_test,y_train,y_test=train_test_split(transformed_x,transformed_y,test_size=0.2)
```

```python
music.isna().sum() #shows how many missing values are there
```

```python

model.fit(x_train ,y_train)
```

```python
model.score(x_test , y_test)
```
